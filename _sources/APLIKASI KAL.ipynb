#PAGE RANK
page rank adalah algoritma yang di gunakan oleh google penelusuran untuk  menentukan peringkat halaman web dalam hasil mesin pencarian.

PageRank bekerja dengan menghitung jumlah dan kualitas tautan ke suatu halaman untuk menentukan perkiraan kasar tentang seberapa penting situs web tersebut. Asumsi yang mendasarinya adalah bahwa situs web yang lebih penting cenderung menerima lebih banyak tautan dari situs web lain

Dari pengertian diatas masih kurang bila belum ada contoh algortimanya.
##Berikut ini contoh algoritma dari pagerank
- caranya yag pertama buat V0 panjang matriksnya harus sama dengan matriks yang ingin dihitung angka nya bebas dengan syarat jika dijumlah harus sama dengan 1
- lakukan perulangan untuk mengalikan V0 dengan matriks A dan tarh hasil matriksnya di V1
- lakukan hingga V saat iini dengan V selanjutnya matriksnya sama(konvegen = proses pengecekan apakah sama)(equibrium = jika hasil dari konvegen sama)
- jika matriks V nya tidak juga sama maka ada beberapa cara yaitu:
- Cara untuk menghentikan iterasi adlah dengan error toleransi atau dengan pembatasan jumlah iterasi
- pada error dengan cara membuatt batasan perbedaan pada contoh ini batasannya 0,01 semakin kecil maka semakin mendekati kebenaran

##cara menghitung error toleransi

- kurangi matris selanjutanya dengan matriks saat ini simpan hasilnya
- kuadratkan hasilnya dan kemudian total tammbahkan semuanya  
- akarkan total tambah tersebut dan jika hasil dari akar tersebut lebihh besar dari batasan maka lanjutkan perulangan
- jika pakai (sum square error) maka cukup sampai sebelum pengakaran  dan berhenti dan mengecek apakah lebih besar dari batasan atau tidak
- jika pakai (root square error maka lakukan sampai pengakaran)
