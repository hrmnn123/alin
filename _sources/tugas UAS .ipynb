1.	Hitunglah PageRank dari graph berikut

image.png

```python
import networkx as nx
import matplotlib.pyplot as plt

Graph= nx.DiGraph()
edges_list = [('1', '4'), ('4', '1'), ('4', '2'), ('4', '5'),('5', '3'),('5', '4'),('5', '2'),('2', '1'),('2', '3'),('3', '1')]
Graph.add_edges_from(edges_list)

pagerank_scores = nx.pagerank(Graph)

for node, score in pagerank_scores.items():
    print(f"Node {node}: PageRank = {score:.4f}")

post = nx.spring_layout(Graph)
nx.draw(Graph, post, with_labels=True, node_size=1000, node_color='pink')
plt.show()
```

2.	Carilah penyelesaian dari Sistem persamaan berikut  dengan Eliminasi Gauss

##langkah penyelesaian

1. Bentuk Matriks Augmented:
Representasikan sistem persamaan dalam bentuk matriks augmented:

\begin{bmatrix}
    1 &-1&1|3\\
    2& 1& 3|8\\
    6& 0& -1|4
    \end{bmatrix}

2. Normalisasi Baris Pertama:
Gunakan elemen pertama pada baris pertama sebagai pivot, dan buat elemen-elemen di bawahnya menjadi nol.

\begin{bmatrix}
    1 &-1&1|3\\
    0& 3& 1|2\\
    0& 6& -7|-14
    \end{bmatrix}

3. Normalisasi Baris Kedua:
Gunakan elemen kedua pada baris kedua sebagai pivot, dan buat elemen-elemen di bawahnya menjadi nol.

\begin{bmatrix}
    1 &-1&1  |3\\
    0& 1& 1/3|2/3\\
    0& 0& -9|-18
    \end{bmatrix}

4. Normalisasi Baris Ketiga:
Buat elemen ketiga pada baris ketiga menjadi satu.

\begin{bmatrix}
    1 &-1&1  |3\\
    0& 1& 1/3|2/3\\
    0& 0& 1|2
    \end{bmatrix}

5. Substitusi Balik:
Selesaikan sistem persamaan dari bawah ke atas.


```python
import numpy as np

# Definisikan matriks augmented
A = np.array([
    [1, -1, 1, 3],
    [2, 1, 3, 8],
    [6, 0, -1, 4]
], dtype=float)

def gauss_elimination(A):
    n = A.shape[0]

    # Forward Elimination
    for i in range(n):
        # Partial pivoting
        max_row = np.argmax(np.abs(A[i:, i])) + i
        A[[i, max_row]] = A[[max_row, i]]

        # Normalize pivot row
        A[i] = A[i] / A[i, i]

        # Eliminate rows below
        for j in range(i + 1, n):
            A[j] = A[j] - A[i] * A[j, i]

    # Back Substitution
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = A[i, -1] - np.sum(A[i, i + 1:n] * x[i + 1:n])
        x[i] /= A[i, i]

    return x

# Solusi dari sistem persamaan
solution = gauss_elimination(A.copy())
print("Solusi dari sistem persamaan:")
print(solution)
```
